```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
a=data.corr().index
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data[a].corr(),annot=True)
```

```python
data.drop(['talk_time'],axis=1,inplace=True)
```

```python
X=data.iloc[:,:-1]
y=data['price_range']
```

```python
print(X.shape,y.shape)
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
cls=ExtraTreesClassifier()
cls.fit(X_train,y_train)
```

```python
imp=cls.feature_importances_
```

```python
b=pd.Series(imp,index=X.columns)
b.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
lin.fit(X_train,y_train)
lin.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=7)
kn.fit(X_train,y_train)
kn.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
ctree=DecisionTreeClassifier(max_depth=7)
ctree.fit(X_train,y_train)
ctree.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
ran=RandomForestClassifier(n_estimators=400,max_depth=7)
ran.fit(X_train,y_train)
ran.score(X_test,y_test)
```

```python
kn.predict(X_test)
```

```python
y_pred=kn.predict(X_test)
```

```python
plt.plot(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
classification_report(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
